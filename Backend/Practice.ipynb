```python
def removespace(text):
   """Remove the spaces from a single sentences. Example: " Hello" -> "Hello" """
   sp=text.split(' ')
   longest=''
   long=0
   for _ in sp:
      if len(_)>long:
         long=len(_)
         longest=_

   return longest
```
